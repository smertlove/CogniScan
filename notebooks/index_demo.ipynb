```python
import pathlib
import sys

PACKAGE_PATH = pathlib.Path.cwd().parent.resolve()
assert PACKAGE_PATH.exists()

sys.path.append(str(PACKAGE_PATH))
```

```python
import json
```

```python
from CogniScan.encoder import Encoder
from CogniScan.index import RamIndex
```

```python
DATADIR = pathlib.Path.cwd().parent.resolve() / "data"
assert DATADIR.exists()
```

```python
demo_data_path = DATADIR / "abstracts_demo.json"

with open(demo_data_path, "r", encoding="utf-8") as file:
    demo_data = json.load(file)

len(demo_data)
```

```python
texts = [
    entry["abstract"]
    for entry
    in demo_data
]

meta = [None] * len(texts)
```

```python
encoder = Encoder()
```

```python
index = RamIndex()
```

```python
index.build_index(
    texts,
    meta,
    encoder,
)
```

```python
index.search(
    "Травматическое повреждение мозга при употреблении алкогольных напитков",
    encoder,
    k=3
)
```

```python
index.search(
    "Сравнить динамику показателей тромбодинамики до и после лечения в двух группах больных эндогенными психическими расстройствами, получавших комбинированную терапию антипсихотиками и антидепрессантами (группа AD+) и не получавших антидепрессанты (группа AD-). В исследование были включены 110 пациенток в возрасте от 16 до 60 лет, средний возраст 29 лет [22; 35]), которые поступили в состоянии обострени",
    encoder,
    k=3
)
```
